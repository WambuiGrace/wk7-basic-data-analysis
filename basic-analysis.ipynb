### Task 1: Load and Explore the Dataset

Load the .csv file dataset

```python
import pandas as pd

try:
    data = pd.read_csv('IRIS.csv')
    print("Data loaded successfully.")
    print(data.tail())
except FileNotFoundError:
    print("Error: The file 'IRIS.csv' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

Explore the structure of the dataset by checking the data types and any missing values.

```python
data.dtypes
data.isnull().sum()

print('Data types: ', data.dtypes)
print()
print('Missing values: ', data.isnull().sum())
```

### Task 2: Basic Data Analysis

Compute the basic statistics of the numerical columns (e.g., mean, median, standard deviation) using .describe().

```python
data.describe()
```

Perform groupings on a categorical column (for example, species, region, or department) and compute the mean of a numerical column for each group.

```python
data.groupby('species').mean()
```

### Task 3: Data Visualization

Line Chart

```python
import matplotlib.pyplot as plt

plt.plot(data.index, data['sepal_length'], label='Sepal Length', color='green')
plt.title('Sepal Length Over Time')
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.show()
```

Bar chart

```python
avg_petal_length = data.groupby('species')['petal_length'].mean()
avg_petal_length.plot(kind='bar', color='green')
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length')
plt.show()
```

Histogram

```python
data['sepal_length'].plot(kind='hist', bins=20, color='green', edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

Scatter plot

```python
plt.scatter(data['sepal_length'], data['petal_length'], color='blue', alpha=0.5)
plt.title('Sepal length vs Petal length')
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.show()
```
